## Import the relevant libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

sns.set()
```

## Load the data

Load data from the csv file: <i> 'Countries.csv'</i>.


```python
data = pd.read_csv('Countries-exercise.csv')
data
```

## Plot the data

Plot the <i>'Longtitude'</i> and <i>'Latitude'</i> columns. 

```python
sns.scatterplot(data, x='Latitude', y='Longitude')
plt.ylim(-180,180)
plt.xlim(-90,90)
plt.show()
```

## Select the features

Create a copy of that data and remove all parameters apart from <i>Longitude</i> and <i>Latitude</i>.

```python
cluster_data = data.copy()
cluster_data.drop('name', axis=1, inplace=True)
cluster_data
```

## Clustering

Assume there are only two clusters. 

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(cluster_data)
```

### Clustering Resutls

```python
preds = kmeans.predict(cluster_data)
preds
```

```python
data['pred'] = preds
data
```

Did you remember to use the <i> c </i> parameter to separate the data by the clusters we defined?

```python
sns.scatterplot(data, x='Latitude', y='Longitude', hue='pred')
```

If you haven't, go back and play around with the number of clusters. 

Try 3, 7 and 8 and see if the results match your expectations!

```python
kmeans = KMeans(n_clusters=7)
kmeans.fit(cluster_data)
preds_7 = kmeans.predict(cluster_data)
preds_7
```

```python
data['preds'] = preds_7
data
```

```python
sns.scatterplot(data, x='Latitude', y='Longitude', hue='preds', palette='tab10')
```
