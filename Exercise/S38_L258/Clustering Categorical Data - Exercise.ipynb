## Import the relevant libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

## Load the data

Load data from the csv file: <i> 'Categorical.csv'</i>.

```python
data = pd.read_csv('Categorical.csv')
data
```

## Map the data

Use the <i>'continent'</i> category for this analysis.

```python
data['continent'] = data['continent'].map({'North America':0,'Europe':1,'Asia':2,'Africa':3,'South America':4, 'Oceania':5,'Seven seas (open ocean)':6, 'Antarctica':7})
data
```

## Select the features

```python
x = data[['continent']]
```

## Clustering

Use 4 clusters initially.

```python
kmeans = KMeans(n_clusters=data['continent'].nunique())
kmeans.fit(x)
```

## Clustering results

```python
preds = kmeans.fit_predict(x)
```

```python
data['preds'] = preds
data
```

## Plot the data

```python
plt.scatter(y=data['Latitude'], x=data['Longitude'], c=data['preds'], cmap='rainbow')
plt.show()
```
